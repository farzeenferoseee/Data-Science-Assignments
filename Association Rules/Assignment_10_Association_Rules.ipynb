```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
#Load dataset
df = pd.read_excel("Online retail.xlsx" , header=None)
```

```python
df.head()
```

#Pre-processing

```python
df[0].isnull().sum()
```

```python
df.duplicated().sum()
```

No null values are available. Since 2325 duplicates are indicated, they are probably not genuine duplicates. Considering it is a retail dataset people may have bought the same items at different days or times. Hence we are not removing them.

```python
#Split the Comma-Separated Strings and convert to Transaction format
transactions = []
for index, row in df.iterrows():
    items_string = row.iloc[0]  # Select the first column (index 0) using iloc
    items_list = [item.strip() for item in items_string.split(',')]  # Remove leading/trailing whitespace
    transactions.append(items_list)

transactions
```

```python
te = TransactionEncoder() #Instantiate TransactionEncoder
te_ary = te.fit(transactions).transform(transactions) # fit and transform the transactions
df_encoded = pd.DataFrame(te_ary, columns=te.columns_) #Create the dataframe
df_encoded
```

```python
#Apriori Algorithm
frequent_itemsets = apriori(df_encoded, min_support=0.03, use_colnames=True)
```

```python
frequent_itemsets
```

```python
frequent_itemsets_sorted = frequent_itemsets.sort_values(by="support", ascending=False)
frequent_itemsets_sorted
```

We have applied a support of 3%. The support values indicate that there is about a 2% chance that the customer will buy mineral water. Items with >1% chance are eggs, spagetti, french fries, chocolate, green tea, and milk.

```python
confidence_rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.04, num_itemsets=len(frequent_itemsets))
confidence_rules_sorted = confidence_rules.sort_values(by="confidence", ascending=False)
confidence_rules_sorted
```

Weak confidence of 0.04-0.05 indicates that 4% to 5% of the transactions containing the antecedent also contain the consequent. In other words, when the antecedent occurs, the consequent *usually* does *not* occur.

When ground beef is bought, there is a 4% confidence that mineral water will also be bought. Similarly, when gound beef is bought spagetti is also bought.

Chocolate is the antescednt for mineral water, spagetti, french fries, and eggs.

```python
#Generate rules
lift_rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1, num_itemsets=len(frequent_itemsets) )

lift_rules_sorted = lift_rules.sort_values(by="lift", ascending=False)
lift_rules_sorted
```

Lift values > 2 is indicated by:
spagetti -> ground beef
ground beef -> spagetti


ground beef and mineral water, frozen veggies and mineral water, spagetti and milk, milk and mineral water, chocolate and milk, pancakes and mineral water, spagetti and mineral water, spagetti and chocolate, chocolate and mineral water, eggs and milk, french fries and chocolate are related to each other among other items. These items can be kept in different areas of the store so customers are encouraged to walk around.

Mineral water is present as a consequent in many cases, so it might be a good idea to keep mineral water close to the checkout counter where it is easy to add to the cast. Alternatively, if antecedent is kept near the beginning of the store, mineral water can be kept at the end of the store so that people are more likely to browse through the store while getting the water.

Chocolate can be kept near the frozen section where milk and french fries are kept. The chocolate section can be opposite to the eggs section.

These essentials that people are buying should be kept in the middle or back of the store so that they browse other items while walking there.

Mineral water prices can be increased while providing a sale on all the other items (or vice versa).



>  What is lift and why is it important in association rules?

Lift measures the strength of the association between two items, considering how frequently each item appears in the dataset. Lift helps us understand if the presence of one item (the antecedent) truly increases the likelihood of another item (the consequent) being present, beyond what we'd expect by chance. A lift value greater than 1 suggests that the items are positively associated, meaning they are more likely to be bought together than independently. The higher the lift, the stronger the association. For example, in the code, the lift values that are greater than 2 indicate that customers are buying these items together more often than it would happen by random chance. The lift values are used to make informed business decisions such as store layout or to put items on sale.

> What is support and confidence? How do you calculate them?

Support measures how frequently an item or a set of items (an itemset) appears in the entire dataset. It's essentially the popularity of the itemset. Support is calculated as the number of transactions containing the itemset divided by the total number of transactions.


Confidence measures the likelihood that the consequent item will be purchased given that the antecedent item has already been purchased. In other words, it measures the strength of the implication of one item to another. Confidence is calculated as the support of the itemset (both antecedent and consequent) divided by the support of the antecedent alone.

> What are some limitations or challenges of association rules mining?

Large Datasets: Association rule mining can be computationally expensive when dealing with very large datasets. The algorithm needs to go through many combinations of items.

Many Rules: Applying association rules to even medium-sized datasets will lead to the production of a huge amount of rules. Many rules that are generated may not be actionable or even interesting.

Setting Thresholds: Choosing appropriate values for support and confidence thresholds can be difficult. If the thresholds are too high, many potentially interesting rules might be missed. If the thresholds are too low, a large number of insignificant rules might be generated.

Spurious Associations: Association rules might find relationships that are statistically significant but not practically meaningful. This is why lift is important to add to the analysis. Lift will help eliminate associations that are not as valuable.

Data Quality: The quality of the results depends heavily on the quality of the data. Errors in the data or missing data can lead to incorrect or misleading rules.

Static Nature: Association rule mining typically analyzes static data and might not capture changes in purchasing behavior over time.
